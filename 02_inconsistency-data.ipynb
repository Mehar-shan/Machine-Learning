```python
import pandas as pd
```

```python
# make pandas dataframe date ,country,  name, sales 2020 ,sales 2021 includes this columns 
df = pd.DataFrame({'date': ['2020-01-01', '2020/02/01', '02-03-2020', '2020-04-01', '2020-05-01'],
    'country': ['US', 'Amrica', 'united states', 'US', 'US'],
    'name': ['john doe', 'doe john', 'jane doe', 'londi doe', 'john masha'],
    'sales_2020': [100, 200, 300, None, 500],
    'sales_2021': [None, 220, 330, 440, 550]})
df 
```

```python
# standrizing date format 
df['date'] = pd.to_datetime(df['date'],errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
# fill NAN with constant value 
df['sales_2020'] = df['sales_2020'].fillna(0)
df.head()
```

```python
# harmonize the name of the country 
country_mapping = {'US': 'UNITED STATES', 'united states': 'UNITED STATES', 'Amrica': 'UNITED STATES'}
df['country'] = df['country'].map(country_mapping)
df.head()
```

```python
#harmonize the name of the name 
name_mapping = {'john doe': 'JOHN DOE', 'doe john': 'JOHN DOE', 'jane doe': 'JOHN DOE', 'londi doe': 'JOHN DOE', 'john masha': 'JOHN DOE'}
df['name'] = df['name'].map(name_mapping)
df.head()
```

```python
# drop duplicates in only name column
df = df.drop_duplicates(subset=['name'])
df.head()
```

```python
#the sales of 2020 will be greater then 2021
#if the condition is not met the remove the row
df = df.drop(df[df['sales_2020'] <= df['sales_2021']].index)
df.head()
```
