```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler , MinMaxScaler, RobustScaler,MaxAbsScaler
```

# standard scalling :


```python
# make example dataframe for age column and weight and height
df = pd.DataFrame({'Age':[20,21,22,23,24,25,26,27,28],'Weight':[50,60,70,80,90,100,110,120,130], 'Height':[150,160,170,180,190,200,210,220,230]})
df
```

```python
#import the scaler
scaler = StandardScaler()

#fit and transform the data
df[['Age', 'Weight', 'Height']] = scaler.fit_transform(df[['Age', 'Weight', 'Height']])

df
```

```python
#import the scaler
scaler = MinMaxScaler()

#fit and transform the data
df[['Age', 'Weight', 'Height']] = scaler.fit_transform(df[['Age', 'Weight', 'Height']])

df
```

```python
#import the scaler
scaler =MaxAbsScaler()

#fit and transform the data
df[['Age', 'Weight', 'Height']] = scaler.fit_transform(df[['Age', 'Weight', 'Height']])

df
```

```python
#import the scaler
scaler = RobustScaler()

#fit and transform the data
df[['Age', 'Weight', 'Height']] = scaler.fit_transform(df[['Age', 'Weight', 'Height']])

df
```
