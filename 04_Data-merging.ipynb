```python
import pandas as pd 
import numpy as np 
```

```python
# make dataframe-01 that consist of id column and randome age and name column
df01 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'name': ['john', 'londi', 'doe', 'jane', 'masha'], 'age': [20, 21, 22, 23, 24]})
df01
```

```python
# make dataframe-02 that consist of id and random city coulum and accupation coulumnn
df02 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'city': ['san jose', 'san francisco', 'san jose', 'san jose', 'san jose'], 'accupation': ['engineer', 'doctor', 'doctor', 'engineer', 'engineer']})
df02
```

```python
# merging df-01 and df-02
merge_df = pd.merge(df01, df02,on='id', how='inner')
merge_df.head()
```

```python
# now concat df-01 and df-02
concat_df = pd.concat([df01, df02], axis=1)
concat_df.head()
```
