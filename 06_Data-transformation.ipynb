```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler , MinMaxScaler, RobustScaler,MaxAbsScaler
```

```python
#generate non normal data (exponential distribution)
np.random.seed(0)
df=np.random.exponential(size=1000)
df=pd.DataFrame(df,columns=['values'])
df.head() 
```

```python
df.shape
```

```python
# make histogram
sns.displot(df['values'],kde=True)
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer
pt_boxpox = PowerTransformer(method='box-cox')
pt_johson = PowerTransformer(method='yeo-johnson')
pt_quantile = QuantileTransformer(output_distribution='normal')
df['Box_cox'] = pt_boxpox.fit_transform(df[['values']] +1)
df['yeo_johson'] = pt_johson.fit_transform(df[['values']] )
df['quantile'] = pt_quantile.fit_transform(df[['values']] )
df.head()

```

```python
# create histogram for all coulumns using sns.hist and kde is true
for col in df.columns:
    sns.displot(df[col],kde=True)
    plt.show()
```
