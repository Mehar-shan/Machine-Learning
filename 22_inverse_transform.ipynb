# inverse transformation  
`inverse transformation` is the process of converting the transformed data back to its original form.
`Label encoder`
`ordinalEncder`
`one hot encoder`

```python
# import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data set
df = sns.load_dataset('titanic')
df.head()
```

```python
# impute misssing values for age , embarked and embark_town and deck 
df['age'].fillna(df['age'].median(),inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
df.drop('deck',axis=1,inplace=True)
df.head()
```

```python
# encode all catigorical varaibles and object varaibles
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_embark_town = LabelEncoder()
le_alive = LabelEncoder()
le_class = LabelEncoder()
le_who = LabelEncoder()
df['sex'] = le_sex.fit_transform(df['sex'])
df['embarked'] = le_embarked.fit_transform(df['embarked'])
df['embark_town'] = le_embark_town.fit_transform(df['embark_town'])
df['alive'] = le_alive.fit_transform(df['alive'])
df['class'] = le_class.fit_transform(df['class'])
df['who'] = le_who.fit_transform(df['who'])
df.head()
```

```python
# now inverse transform
df['sex'] = le_sex.inverse_transform(df['sex'])
df['embarked'] = le_embarked.inverse_transform(df['embarked'])
df['embark_town'] = le_embark_town.inverse_transform(df['embark_town'])
df['alive'] = le_alive.inverse_transform(df['alive'])
df['class'] = le_class.inverse_transform(df['class'])
df['who'] = le_who.inverse_transform(df['who'])
df.head()
```

# Same Procedure for one hot encoding

```python
# import dataset
df_02 = sns.load_dataset('tips')
df_02.head()
```

```python
# encode catigorical  columns using one hot encoding
# assume df_02 is your original DataFrame

# one-hot encode the columns
ohe_sex = OneHotEncoder()
ohe_smoker = OneHotEncoder()
ohe_day = OneHotEncoder()
ohe_time = OneHotEncoder()

encoded_sex = ohe_sex.fit_transform(df_02['sex'].values.reshape(-1,1)).toarray()
encoded_smoker = ohe_smoker.fit_transform(df_02['smoker'].values.reshape(-1,1)).toarray()
encoded_day = ohe_day.fit_transform(df_02['day'].values.reshape(-1,1)).toarray()
encoded_time = ohe_time.fit_transform(df_02['time'].values.reshape(-1,1)).toarray()

df_02.head()
```

```python
# create a new DataFrame with the encoded values
encoded_df = pd.DataFrame({
    'sex_0': encoded_sex[:, 0],
    'sex_1': encoded_sex[:, 1],
    'smoker_0': encoded_smoker[:, 0],
    'smoker_1': encoded_smoker[:, 1],
    'day_0': encoded_day[:, 0],
    'day_1': encoded_day[:, 1],
    'time_0': encoded_time[:, 0],
    'time_1': encoded_time[:, 1]
})

# inverse transform the encoded values
# inverse transform the encoded values
# inverse transform the encoded values
inverse_transformed_df = pd.DataFrame({
    'sex': ohe_sex.inverse_transform(encoded_df[['sex_0', 'sex_1']].values).squeeze(),
    'smoker': ohe_smoker.inverse_transform(encoded_df[['smoker_0', 'smoker_1']].values).squeeze(),
    'day': encoded_df[['day_0', 'day_1']].values.argmax(axis=1),
    'time': encoded_df[['time_0', 'time_1']].values.argmax(axis=1)
})
df_02.head()
```
