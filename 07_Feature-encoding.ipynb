```python

df.df= sns.load_dataset('tips')
df.head()
```

```python
df['time'].value_counts()
```

```python
# lets encode time through lable encoding using skllearn

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder

le=LabelEncoder()
df['endoded-time']=le.fit_transform(df['time'])
df.head()
```

```python
df['day'].value_counts()
```

```python
# using ordinal encoding the day column 

oe=OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded-day']=oe.fit_transform(df[['day']])
df.head()
```

```python
df['encoded-day'].value_counts()
```

```python
#now one hot encoding the day column
ohe=OneHotEncoder()
ohe.fit_transform(df[['sex']]).toarray()
```
