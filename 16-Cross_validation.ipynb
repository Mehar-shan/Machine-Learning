```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# load tip dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# define the features and target
X = df[['total_bill','tip', 'size']]
y = df['sex']
```

```python
# define tha Gaussian naive bayes model
model = GaussianNB()
```

```python
# perform cross validation with k=5
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
# print the score for each fold and the mean score
print("Scores:", scores)
print("Mean score:", scores.mean())
print("Standard deviation:", scores.std())
```

```python
# deife decsion tree model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
# perform cross validation with k=5
scores = cross_val_score(model, X, y, cv=10, scoring='accuracy')
# print the score for each fold and the mean score
print("Scores:", scores)
print("Mean score:", scores.mean())
print("Standard deviation:", scores.std())
```

```python
#  plot the accuracy for each fold using barplot
import matplotlib.pyplot as plt
#  plot the accuracy for each fold using barplot,plot the results using plt.plot and use diffirent colors for each model
plt.figure(figsize=(8,8))
plt.plot(range (1,11),scores,color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Accuracy for each fold')
plt.show()
```
