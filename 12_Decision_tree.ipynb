# impurity, entropy and information gain
`impurity, entropy and information gain` the three measures of impurity are entropy, gini impurity and information gain.use for decision tree.


```python
import math
```

```python
# example data set , with  two classes A and B
# suppose in datasets of 10 elements, 4 are class A  and 6 are class B

A = 4   
B = 6
total = A + B


```

```python
# lets calculate proportions 

prop_A = A/total
prop_B = B/total

print(prop_A)
print(prop_B)
```

```python
# entropy calculate 
#entropy is measure of uncertainty

entropy = -(prop_A * math.log(prop_A,2) + prop_B * math.log(prop_B,2))
print(entropy)
```

```python
# gini impurity 
# gini impurity is measure of missclassification

gini = 1 - (prop_A**2 + prop_B**2)
print(gini)
```

```python
# calculating information gain with complete procedure 

# information gain is measure of how much information is gained by partitioning the dataset into two parts

information_gain = entropy - ((A/total) * entropy + (B/total) * entropy)
print(information_gain)


```

# Decsion Tree
`Decsion Tree` is a supervised learning algorithm that uses a flowchart-like structure to recursively partition a dataset based on the values of a feature, until a pure class is encountered.

```python
# import libaraies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')
```

```python
# import titanic data set
df = sns.load_dataset("titanic")
df.head()
```

```python
# drop deck column
df.drop('deck',axis=1,inplace=True)
df.head()
```

```python
# impute missing value of age and fare with median 

imputer = SimpleImputer(strategy='median')
df['age'] = imputer.fit_transform(df[['age']])
df['fare'] = imputer.fit_transform(df[['fare']])
df.head()
# impute missing value of embark and embark town with mode

df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
# encode the catigorical and object variable using for loop and label encoding

for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype == 'category':
        df[col] = LabelEncoder().fit_transform(df[col])
df.head()
```

```python
# split the data into x and y

x = df.drop(['survived','alive'],axis=1)
y = df['survived']
```

```python
# split the data into train x and y

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# create and train the model with pres 

model = DecisionTreeClassifier(criterion='entropy',random_state=42,max_depth=6)
model.fit(x_train,y_train)
# pridict 
y_pred=model.predict(x_test)
```

```python
# evaluate model 
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# save the descion tree
from sklearn.tree import export_graphviz

export_graphviz(model,out_file='tree.dot',feature_names=x.columns,filled=True)
```
